```python
import matplotlib.pyplot as plt
%matplotlib inline


x= [2,3,5,8]
labels=["A","B","U","H"]
explode= [0,0,0.2,0]
colors=["blue","pink","yellow","turquoise"]
wedge_properties={"edgecolor":"k",'linewidth':1}




plt.title("color chart")
plt.pie(x,labels=labels,explode=explode,colors=colors,wedgeprops=wedge_properties,autopct="%1.1f%%",pctdistance=0.7)
plt.show()

```
